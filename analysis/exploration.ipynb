# Exploracion de base datos!!

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Cargando la base de datos
# https://www.kaggle.com/miroslavsabo/young-people-survey
base = pd.read_csv('responses.csv')
#Eliminando casos con NAs
base1 = base.dropna(axis=0,how='any')
print len(base), len(base1)
base1.head()
```

```python
#Descriptivos basicos
base1.describe()
```

## Analisis de las variables musicales. (de la 1 a la 19)

```python
# Matriz de correlaciones
corr = base1.iloc[:,0:19].corr()
corr
```

```python
# Grafico gonito de correlaciones
corr = base1.iloc[:,0:19].corr()

%pylab inline
plt.figure(figsize=(15, 15))
plt.imshow(corr, cmap='RdYlGn', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns);
plt.suptitle('Correlations Heat Map', fontsize=15, fontweight='bold')
plt.show()
```
